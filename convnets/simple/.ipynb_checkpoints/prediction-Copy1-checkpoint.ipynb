```python
from tensorflow.keras.models import load_model
import cv2
import tensorflow as tf
import numpy as np
import os
model = load_model('first_model.h5')

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
val_cat_dir = '/home/user/Документы/convnets/raw-img/cats/'
val_cow_dir = '/home/user/Документы/convnets/data/validation/cows/'
error_counter = 0
images = os.listdir(val_cow_dir)

    
for name in images:
    img = cv2.imread(val_cow_dir + name)
    img = cv2.resize(img,(150,150))
    img = np.reshape(img,[1,150,150,3])
    img = tf.cast(img, tf.float32)
    
    classes = model.predict_classes(img)
    if classes[0][0] < 0.01:
        error_counter = error_counter + 1
        info = str(error_counter) + "/" + str(len(images))
        print(info)

```
