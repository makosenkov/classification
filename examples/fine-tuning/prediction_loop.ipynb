```python
from tensorflow.keras.models import load_model
import cv2
import tensorflow as tf
import numpy as np
import os
import tensorflow.keras.preprocessing.image as image
from tensorflow.keras import applications

model = load_model('/home/models/fine-tuned/bottleneck_fc_model.h5')

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
val_cat_dir = '/home/user/Документы/convnets/raw-img/cats/'
val_but_dir = '/home/user/Документы/convnets/raw-img/butterflies/'
val_cow_dir = '/home/user/Документы/convnets/raw-img/cows/'
val_chicken_dir = '/home/user/Документы/convnets/raw-img/chicken/'
val_dog_dir = '/home/user/Документы/convnets/raw-img/dogs/'

img_width, img_height = 150, 150

model1 = applications.VGG16(include_top=False, weights='imagenet')
error_counter = 0
zero_counter = 0
samples_count = 501
for i in range(1,samples_count):
    img = image.load_img(val_chicken_dir + 'chicken' + str(i) + '.jpg', target_size=(img_width, img_height))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    features = model1.predict(x)

    # this is where you get your prediction
    prediction = model.predict(features)
    if prediction[0][2] == 0:
        error_counter += 1
        if (prediction[0][0] == 0 and
            prediction[0][1] == 0 and
            prediction[0][2] == 0 and
            prediction[0][3] == 0 and
            prediction[0][4] == 0):
            zero_counter += 1

    if (error_counter != 0 and (i % 20 == 0)):
        print('mistakes rate: ' + str(1 - error_counter / samples_count))
        print('zeros rate:    ' + str(1 - zero_counter / samples_count))
print('done')
```
